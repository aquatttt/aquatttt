# 4.스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun gothic'  #한글폰트 설정해야함
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시 - 깨지는 현상 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, linewidth=5)
```

```python
plt.plot(x, y, marker='*')
```

# Marker

```python
plt.plot(x, y, marker='*', linestyle='None')
```

```python
plt.plot(x, y, marker='v')
```

```python
plt.plot(x, y, marker='v', markersize=10)
```

```python
plt.plot(x, y, marker='X', markersize=10)
```

```python
plt.plot(x, y, marker='X', markersize=10)
```

```python
plt.plot(x, y, marker='o', markersize=20, markeredgecolor= 'r')
```

```python
plt.plot(x, y, marker='o', markersize=20, markeredgecolor= 'r', markerfacecolor='yellow')
```

# 선스타일

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, linestyle='--') # 두꺼워진 점선
```

# 색깔

```python
plt.plot(x, y, color='#ff0000')
```

```python
plt.plot(x, y, color='g')
```

# 포맷

```python
plt.plot(x, y, 'ro--')
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, marker='o', mfc='red', mec='blue', ls=":")
```

# 투명도

```python
plt.plot(x, y, marker='o', mfc='red', mec='blue', ls=":", alpha=0.3) # alpha는 투명도
```
