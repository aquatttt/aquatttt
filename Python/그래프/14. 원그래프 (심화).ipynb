# 원그래프 (심화)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun gothic'  #한글폰트 설정해야함
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시 - 깨지는 현상 해결
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
colors = ['b','g','y','c','m','r']
explode = [0.05] * 6
wedgeprops ={'width':0.6}
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False, explode = explode, colors=colors, wedgeprops = wedgeprops)
plt.show()
```

```python
wedgeprops ={'width':0.6}
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, wedgeprops = wedgeprops)
plt.show()
```

```python
def custom_autopct(pct):
    #return ('%.1f%%' % pct) if pct >= 10 else ''
    #return '{:,1f}%'.format(pct) if pct >= 10 else ''
    return '{:.0f}%'.format(pct) if pct >= 10 else '' # 소수점 제외 정수
plt.pie(values, labels=labels, autopct=custom_autopct, startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops, pctdistance=0.7)
plt.show()
```
