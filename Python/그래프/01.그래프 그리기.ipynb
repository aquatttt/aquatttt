```python
import matplotlib.pyplot as plt
```

# 그래프 기본

```python
x = [1, 2, 3]
y = [2, 4, 8]
plt.plot(x,y)
```

```python
print(plt.plot(x,y))
```

```python
plt.plot(x,y)
plt.show()
```

## 그래프 제목(title) 설정

```python
plt.plot(x,y)
plt.title('Line Graph')
```

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun gothic'  #한글폰트 설정해야함
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시 - 깨지는 현상 해결
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist                         # 사용가능 폰트확인
[f. name for f in fm.fontManager.ttflist]    #폰트 이름만 출력
```

```python
plt.plot(x,y)
plt.title('꺽은선 그래프') #한글폰트 설정해야함
```
