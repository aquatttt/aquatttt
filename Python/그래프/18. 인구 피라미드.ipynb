# 데이터 시각화


```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun gothic'  #한글폰트 설정해야함
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시 - 깨지는 현상 해결
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 추출
male_data = df_m.iloc[0] // 1000
female_data = df_w.iloc[0] // 1000

# y 축 범위 설정
y = np.arange(len(df_w.columns))

# 그래프 그리기
fig, ax = plt.subplots(figsize=(10, 7))
ax.barh(y, male_data, color='blue', alpha=0.7, label='Male')
ax.barh(y, -female_data, color='pink', alpha=0.7, label='Female')  # 음수로 그려서 반대편에 위치시킴

# 축과 레이블 설정
ax.set_yticks(y)
ax.set_yticklabels(df_w.columns)
ax.invert_yaxis()  # y 축 순서 뒤집기
ax.legend(loc='upper right')
ax.set_xlabel('Population (thousands)')

# 그래프 보여주기
plt.show()
```
