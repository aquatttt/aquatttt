# 여러 그래프

```python
import pandas as pd
df = pd.read_excel('../../pandas/score.xlsx')
df
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun gothic'  #한글폰트 설정해야함
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시 - 깨지는 현상 해결
```

```python
fig,axs = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('여러 그래프 넣기')

# 1번 그래프
axs[0, 0]. bar (df['이름'], df['국어'], label='국어점수')
axs[0, 0]. set_title('첫 번째 그래프')
axs[0, 0]. legend()
axs[0, 0]. set(xlabel = '이름', ylabel='점수')
axs[0, 0]. set_facecolor('lightyellow')
axs[0, 0]. grid(linestyle='--', linewidth=0.5)

# 2번 그래프
axs[0, 1 ].plot(df['이름'], df['수학'], label='수학')
axs[0, 1 ].plot(df['이름'], df['영어'], label='영어')
axs[0, 1 ].legend()

# 3번 그래프
axs[1, 0 ].plot(df['이름'], df['키'])

# 4번 그래프
axs[1, 1 ].plot(df['이름'], df['사회'], color='g', alpha=0.6)
```
