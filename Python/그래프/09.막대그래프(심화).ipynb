# 막대 그래프(심화)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun gothic'  #한글폰트 설정해야함
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시 - 깨지는 현상 해결
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.barh(labels, values)
plt.xlim(175,195)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('.')
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)

for idx, rect in enumerate(bar):  # idx는 0부터 시작, rect 각 bar에대한 크기 정보
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha='right', color='blue')
```
