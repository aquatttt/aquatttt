# COVID-19 백신 종류별 접종 인구

```python
days = [1, 2, 3]
Az = [2, 4, 8] # 단위:만명 1일~3일까지 아스트라제네카
pfizer = [5, 1, 3] # 화이자
moderna = [1, 2, 5] # 모더나

plt.plot(days, Az)
plt.plot(days, pfizer)
plt.plot(days, moderna)
```

```python
plt.plot(days, Az, label='Az')
plt.plot(days, pfizer, label='pfizer', marker='o', ls='--')
plt.plot(days, moderna,label='morderna', marker='s', ls='-')

plt.legend()
```

```python
plt.plot(days, Az, label='Az')
plt.plot(days, pfizer, label='pfizer', marker='o', ls='--')
plt.plot(days, moderna,label='morderna', marker='s', ls='-')

plt.legend(ncol=3)
```
